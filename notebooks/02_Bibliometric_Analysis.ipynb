```python
# ==========================================================
# Read processed dataset
# ==========================================================

filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print("="*60)
print("Processed dataset loaded successfully")
print("="*60)

print(f"\nDocuments: {df.shape[0]}")
print(f"Variables: {df.shape[1]}")

display(df.head())
```

```python
# ==========================================================
# General bibliometric indicators
# ==========================================================

print("="*60)
print("GENERAL BIBLIOMETRIC INDICATORS")
print("="*60)

print(f"\nTotal publications: {len(df)}")

print(f"\nPublication period: {df['Year'].min()} - {df['Year'].max()}")

print(f"\nDocument types:")
print(df["Document Type"].value_counts())

print(f"\nLanguages:")
print(df["Language of Original Document"].value_counts())

print(f"\nAverage citations per publication:")
print(round(df["Cited by"].mean(),2))
```
