```python
df = pd.read_csv("processed_scopus.csv")

print("="*60)
print("Dataset loaded successfully")
print("="*60)

print(f"Documents: {len(df)}")
```

```python
publications = (
    df.groupby("Year")
      .size()
      .reset_index(name="Publications")
)

publications
```

```python
fig = px.line(

    publications,

    x="Year",

    y="Publications",

    markers=True,

    title="Scientific Production by Year"

)

fig.show()
```

```python
language = (

    df["Language of Original Document"]

    .value_counts()

    .reset_index()

)

language.columns = [

    "Language",

    "Documents"

]

language
```

```python
fig = px.pie(

    language,

    values="Documents",

    names="Language",

    title="Language Distribution"

)

fig.show()
```

```python
doctype = (

    df["Document Type"]

    .value_counts()

    .reset_index()

)

doctype.columns=[

    "Document",

    "Quantity"

]

doctype
```

```python
fig = px.bar(

    doctype,

    x="Document",

    y="Quantity",

    title="Document Types"

)

fig.show()
```

```python
sentiment = pd.DataFrame({

    "Sentiment":[
        "Neutral",
        "Positive",
        "Negative"
    ],

    "Documents":[
        343,
        123,
        17
    ]

})

sentiment
```

```python
fig = px.bar(

    sentiment,

    x="Sentiment",

    y="Documents",

    color="Sentiment",

    title="Sentiment Analysis Distribution"

)

fig.show()
```

```python
annual_sentiment = pd.DataFrame({

    "Year":[
        2018,2019,2020,2021,
        2022,2023,2024,2025,2026
    ],

    "Score":[
        0.000000,
        0.500000,
        0.200000,
        0.045455,
        0.148148,
        0.297297,
        0.328767,
        0.243750,
        0.143885
    ]

})

annual_sentiment
```

```python
fig = px.line(

    annual_sentiment,

    x="Year",

    y="Score",

    markers=True,

    title="Average Sentiment by Year"

)

fig.show()
```

```python
forecast = pd.DataFrame({

    "Year":[
        2026,
        2027,
        2028,
        2029,
        2030
    ],

    "Prediction":[
        128.80,
        147.58,
        166.37,
        185.15,
        203.93
    ]

})

forecast
```

```python
fig = px.line(

    forecast,

    x="Year",

    y="Prediction",

    markers=True,

    title="Forecast of Scientific Publications"

)

fig.show()
```
